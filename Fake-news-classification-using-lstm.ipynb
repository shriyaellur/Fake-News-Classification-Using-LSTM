# **Fake News Classification using LSTM**

## **1.Text Preprocessing**

```python
import pandas as pd
import numpy as np
```

```python
data_true = pd.read_csv('../input/fake-and-real-news-dataset/True.csv')
data_fake = pd.read_csv('../input/fake-and-real-news-dataset/Fake.csv')

data_true['label'] = 1
data_fake['label'] = 0
data = pd.concat([data_true, data_fake], axis = 0)
```

```python
data.shape
```

```python
data.head()
```

```python
df=data.drop(['title','subject','date'],axis=1)
df.head()
```

```python
df.dtypes
```

```python
df=df.reset_index(drop=True)
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
df['text']=df['text'].apply(str)
```

```python
ps=PorterStemmer()
corpus=[]
for i in range(0,len(df)):
  rev=re.sub('[^a-zA-Z]',' ',df['text'][i])
  rev=rev.lower()
  rev=rev.split()
  
  rev=[ps.stem(word) for word in rev  if not word in stopwords.words('english')]
  rev=' '.join(rev)
  corpus.append(rev)
```

```python
(corpus[:1])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
X=cv.fit_transform(corpus).toarray()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
X=pad_sequences(X,maxlen=300)
```

```python
y=df['label']
```
